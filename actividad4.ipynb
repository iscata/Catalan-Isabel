```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
#Funciones para leer el archivo csv y cargar las columnas
def cargar_dataset(ruta_archivo:str) -> pd.DataFrame:
    df = pd.read_csv(ruta_archivo)
    return df

def cargar_columnas (df: pd.DataFrame, columnas:list)  -> pd.DataFrame:
    return df[columnas] 

def cargar_datos(ruta_archivo: str, columnas_deseadas: list):
    dataset = cargar_dataset(ruta_archivo)
    X = dataset[columnas_deseadas].values
    scaler = MinMaxScaler()
    X_escalado = scaler.fit_transform(X)
    return X_escalado, dataset, scaler

def aplicar_kmeans(X_escalado, num_clusters=6, random_state=42):
    kmeans = KMeans(n_clusters=num_clusters, random_state=random_state) 
    kmeans.fit(X_escalado)  
    centroides = kmeans.cluster_centers_  
    clusters_asignados = kmeans.labels_  
    inercia = kmeans.inertia_ 
    return kmeans, centroides, clusters_asignados, inercia

def generar_reporte(KMeans, centroides, clusters_asignados, inercia):
    print(f':\n{centroides}\n')
    print(f'Inercia: {inercia}')
    print(f'Clusters: {clusters_asignados}\n')
    
def generar_visualizaciones(X_escalado, kmeans, centroides):
    plt.figure(figsize=(8, 6))
    plt.scatter(X_escalado[:, 0], X_escalado[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.6, label="Puntos")
    plt.scatter(centroides[:, 0], centroides[:, 1], c='blue', marker='X', s=200, label="Centroides")
    plt.title('Distribución de los puntos y los centroides')
    plt.xlabel()
    plt.ylabel()
    plt.show()
```

```python
#Selección de ambas variables desde las columnas del archivo csv 
ruta_archivo = 'kmeans.1.csv'
dataset = cargar_dataset(ruta_archivo)

columnas_deseadas = ['A', 'B']

datos_xy = cargar_columnas(dataset, columnas_deseadas)
print(datos_xy.head())
```

```python
#función de análisis de centroides
def analizar_centroides(kmeans, scaler, columnas_deseadas):
    centroides_originales = scaler.inverse_transform(kmeans.cluster_centers_)
    df_centroides = pd.DataFrame(centroides_originales, columns=columnas_deseadas)
    print('Reporte de Centroides en la Escala Original:')
    print(df_centroides)
```
